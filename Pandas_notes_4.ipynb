```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets\\weather_data2.csv')
```

```python
df
```

<h2>Replacing invalid values with NaN</h2>

```python
new_df = df.replace(-99999, np.NaN)
```

```python
new_df
```

<h2>Replacing special or invalid values for each column for their uniqueness for each column</h2>

```python
new_df = df.replace({
    'temperature': -99999,
    'windspeed': -99999,
    'event': 'No Event'
}, np.NaN)
```

```python
new_df
```

```python
df
```

<h2>Replacing missing data with the mapping technique</h2>

```python
new_df = df.replace({
    -99999: np.NaN,
    'No Event': 'Sunny'
})
```

```python
new_df
```
