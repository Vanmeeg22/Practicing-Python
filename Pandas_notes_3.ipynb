<h2>Converting day column from 'str' into date type</h2>

```python
df = pd.read_csv('datasets\\weather_data.csv', parse_dates = ['day'])
```

```python
df
```

```python
type(df.day[0])
```

```python
df.set_index('day', inplace = True)
```

```python
df
```

<h2>Filling NaN values with 0</h2>

```python
df.fillna(0)
```

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event'
})
```

```python
new_df
```

<h2>Filling NA values with forward fill or ffill, bringing forward the previous value.</h2>

```python
new_df = df.fillna(method = 'ffill')
```

```python
new_df
```

<h2>Filling NA values with before fill or bfill</h2>

```python
new_df = df.fillna(method = 'bfill')
```

```python
new_df
```

<h2>Copying the data horizontally into the empty columns</h2>

```python
new_df = df.fillna(method = 'bfill', axis = 'columns')
```

```python
new_df
```

<h2>Copying the data values vertically into the empty values only once by setting a limit</h2>

```python
new_df = df.fillna(method = 'ffill', limit = 1)
```

```python
new_df
```

<h2>Using linear interpolation to fill in missing values</h2>

```python
new_df = df.interpolate()
```

```python
new_df
```

<h2>Performing interpolation by date for more accuracy in data prediction</h2>

```python
new_df = df.interpolate(method = 'time')
```

```python
new_df
```

<h2>Dropping all the rows containing any empty values</h2>

```python
new_df = df.dropna()
```

```python
new_df
```

<h2>Dropping all the rows containing all the values as empty</h2>

```python
new_df = df.dropna(how = 'all')
```

```python
new_df
```

<h2>Dropping all the rows with a certain threshold of number of empty values</h2>

```python
new_df = df.dropna(thresh = 1)  # Will drop all the rows that does not have atleast 1 valid value
```

```python
new_df
```

```python
new_df = df.dropna(thresh = 2)  # will drop all the rows that does not have atleast 2 valid values
```

```python
new_df
```

```python
new_df = df.dropna(thresh = 3) # will drop all the rows that does not have atleast 3 valid values
```

```python
new_df
```
